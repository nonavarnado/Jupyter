## Science Question: 

To investigate if Biogenic Volatile Organic Compounds (BVOCs), primarily
through isoprene, which accounts for 50% of global BVOC emissions, in the gas
phase (not aerosol), are linked to methane (ch4) lifetime. 

This project looks at the ch4 concentrations in UCARâ€™s WACCM2 model experiments. PiClim-VOC is the VOC
baseline against preindustrial SSTs and forcing but with 2014 VOC emissions. 

I will compare that to PiClim-2xVOC, which assumes twice the VOC emissions of
the first model experiment (PiClim-VOC) and see how ch4 compares between the two. This is a very blunt way of investigating very complex chemistry and dynamics. The original idea to explore BVOCs - OH - CH4 interactions through model comparisons was thanks to the paper linked below and by professor Jen Kay's recent ESS seminar in describing an investigation method in cross model comparison and model simplification strategies. 

**NOTE:"Effective radiative forcing from emissions of reactive gases and aerosols - a multi-model comparison." https://acp.copernicus.org/articles/21/853/2021/**
