## 2.1.9 The Planck function 
(Zender, 2010) https://ia600908.us.archive.org/25/items/flooved3325/flooved3325.pdf
describes the intensity of blackbody radiation as a function of temperature and wavelength. This can be expressed as a function with two variables.


$$ B(\lambda, T) = \frac{2 \pi hc^2}{\lambda^5}\cdot \frac{1}{\exp\left(\frac{hc}{\lambda k_B T}\right) - 1} $$





$$ c = 3.0 \times 10^8 ms^{-1} \text{ is the speed of light} $$ 

$$ k_B = 1.380649 \times 10^{−23} JK^{-1} \text{ is the Boltzmann constant} $$ 

$$ h = 6.62607015 \times 10^{−34} J\cdot s \text{ is the Planck constant} $$ 

$$ \lambda \text{ is the wavelength in meters}$$ 

$$ T \text{ is absolute temperature (Kelvin)} $$


```python
#Planck's Law Function (41b)
def B(wave,T):
    c = 3e8
    kB = 1.380649e-23
    h = 6.62607015e-34
    pt1 = 2*np.pi*h*c**2/(wave**5)
    pt2 = 1/(np.exp(h*c/(wave*kB*T))-1)
    return (pt1*pt2)
```

```python
# Create Linspace for value range -- wavelength units for 100nm to 35000nm range to m conversion with output in w
w = np.linspace(100e-9, 35000e-9, 1000)
b_earth = B(w, 288)
b_sun = B(w, 5772)
b_mars = B(w, 210)
```

```python
# plot blackbody radiation - Earth and Mars
ax = py.axes()
ax.plot(w*1e9, b_earth*1e-6);
ax.plot(w*1e9, b_mars*1e-6);

ax.set(xlim = (-100,35000), xlabel = "Wavelength (nanometers)", ylabel = "Power density (watts/m$^2$ per $\mu$m)", 
       title = "Blackbody radiation of Earth and Mars");
```

```python
# plot blackbody radiation - Sun (axis scale issues are better re-plotted)
ax = py.axes()
ax.plot(w*1e9, b_sun*1e-6, color = 'red');
ax.set(xlim = (-100,2500), xlabel = "Wavelength (nanometers)", ylabel = "Power density (watts/m$^2$ per $\mu$m)", title = "Blackbody radiation of the sun");
```
